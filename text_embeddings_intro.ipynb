## Similarity Score

```python

# from openai.embeddings_utils import cosine_similarity   #---> this is no longer available in new version
import numpy as np
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
df['similarity_score'] = df['embedding'].apply(lambda x: cosine_similarity(x, text_embedding))
```

```python
df
```

```python
df.sort_values("similarity_score", ascending=False)
```
